# Al-Driven Enrollment Forecasting & Resource Optimization

## In this, we will create a real world balance dataset using python library "faker"

```python
import pandas as pd
import random
from faker import Faker
```

```python
fake=Faker()
```

```python
# sample one row
record = {
    "course_id": random.randint(100, 200),
    "tuition_fee": random.randint(20000, 100000),
    "waitlist": random.randint(0, 50),
    "retention_rate": round(random.uniform(0.5, 0.95), 2),
    "feedback_score": round(random.uniform(1, 5), 1),
    "faculty_rating": round(random.uniform(2, 5), 1),
    "capacity": random.randint(30, 120),
    "enrollment": random.randint(20, 120)
}

print(record)
```

```python
data = []

for _ in range(5000):
    capacity = random.randint(30, 120)
    waitlist = random.randint(0, 50)
    retention_rate = round(random.uniform(0.5, 0.95), 2)
    feedback_score = round(random.uniform(1, 5), 1)
    
    # ✅ DEFINE FIRST
    faculty_rating = round(random.uniform(2, 5), 1)

    # ✅ THEN USE IT
    enrollment = int(
        0.6 * capacity +
        1.0 * waitlist +
        20 * retention_rate +
        10 * feedback_score +
        8 * faculty_rating +
        random.randint(-3, 3)
    )

    record = {
        "course_id": random.randint(100, 200),
        "tuition_fee": random.randint(20000, 100000),
        "waitlist": waitlist,
        "retention_rate": retention_rate,
        "feedback_score": feedback_score,
        "faculty_rating": faculty_rating,  # reuse variable
        "capacity": capacity,
        "enrollment": max(0, enrollment)
    }

    data.append(record)
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("../data/enrollment_data.csv", index=False)
```
